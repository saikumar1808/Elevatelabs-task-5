# Count how many games of each genre were released each year

```python
genre_trend = df_cleaned.groupby(['Year', 'Genre']).size().unstack().fillna(0)
```

# Plot only the top 5 genres by total releases

```python
top_genres = genre_trend.sum().sort_values(ascending=False).head(5).index
genre_trend[top_genres].plot(figsize=(12,6))
```

# Sum global sales by year and publisher

```python
publisher_sales_by_year = df_cleaned.groupby(['Year', 'Publisher'])['Global_Sales'].sum().reset_index()
```

# Get the top publisher for each year

```python
top_publishers = publisher_sales_by_year.loc[publisher_sales_by_year.groupby('Year')['Global_Sales'].idxmax()]
```

```python
plt.figure(figsize=(14,6))
sns.barplot(data=top_publishers.sort_values('Year'), x='Year', y='Global_Sales', hue='Publisher', dodge=False)

plt.title('Top Publisher by Global Sales Each Year')
plt.xticks(rotation=45)
plt.ylabel('Global Sales (in millions)')
plt.tight_layout()
plt.show()
```
